# Tweets Preprocessing

```python
stop_words = stopwords.words('english')
```

```python
def remove_emoji(string):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)
```

```python
def clean_tweet(tweet):
    if type(tweet) == np.float:
        return ""
    temp = tweet.lower()
    temp = ''.join([i for i in temp if not i.isdigit()])
    temp=temp.replace("@","")
    temp=temp.replace(".","")
    temp=temp.replace(",","")
    temp=temp.replace("`","")
    temp=temp.replace("~","")
    temp=temp.replace("!","")
    temp=temp.replace("#","")
    temp=temp.replace("$","")
    temp=temp.replace("%","")
    temp=temp.replace("^","")
    temp=temp.replace("&","")
    temp=temp.replace("*","")
    temp=temp.replace("(","")
    temp=temp.replace(")","")
    temp=temp.replace("_","")
    temp=temp.replace("-"," ")
    temp=temp.replace("+","")
    temp=temp.replace("=","")
    temp=temp.replace("{","")
    temp=temp.replace("}","")
    temp=temp.replace("[","")
    temp=temp.replace("]","")
    temp=temp.replace("|","")
    temp=temp.replace(":","")
    temp=temp.replace(";","")
    temp=temp.replace("/","")
    temp=temp.replace("?","")
    temp=temp.replace(">","")
    temp=temp.replace("<","")
    temp=re.sub(r'[^\w]', ' ', temp)
    temp=re.sub(r'http\S+', '', temp)
    temp = re.sub("[^a-zA-Z0-9]+", " ",temp)
    if "…" in temp:
        read_ind=temp.index("…")
        temp=temp[:read_ind]
    if "read" in temp:
        read_ind=temp.index("read")
        temp=temp[:read_ind]
    temp=remove_emoji(temp)
    temp = temp.split()
    temp = [w for w in temp if not w in stop_words]
    temp = " ".join(word for word in temp)
    return temp
```

# Scraping Tweets

```python
hashtags=["stocks","nse","bse","nasdaq","s&p500","dowjones","dowfutures","nifty50","sensex","banknifty"]
twe=[]
for hashtag in hashtags:
    search_words = "#"+hashtag
    tweets = tw.Cursor(api.search_tweets,
                  q=search_words, 
                  lang="en",
                  result_type="mixed").items(150)
    for tweet in tweets:
        print(clean_tweet(tweet.text))
        print("******************")
        twe.append(clean_tweet(tweet.text))
```
